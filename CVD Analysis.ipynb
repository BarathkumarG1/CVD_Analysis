```python
# 4.	Identify the data variables which might be categorical in nature. Describe and explore these variables using appropriate tools e.g., count plot. 

numerical_feature=[feature for feature in Cleaned_dataframe.columns if Cleaned_dataframe[feature].dtypes != 'O']
discrete_feature=[feature for feature in numerical_feature if len(Cleaned_dataframe[feature].unique())<7]
continuos_feature=[feature for feature in numerical_feature if feature not in discrete_feature]
categorical_feature=[feature for feature in Cleaned_dataframe.columns if feature not in numerical_feature]
print("numerical features ={}" .format(len(numerical_feature)))
print("discrete features  ={}".format(len(discrete_feature)))
print("contious feature   ={}".format(len(continuos_feature)))
print("catagorical feature ={}".format(len(categorical_feature)))
```

```python
import seaborn 
import matplotlib.pyplot 
```

```python
for feature in discrete_feature:
    seaborn.countplot(Cleaned_dataframe[feature])
    matplotlib.pyplot .xlabel(feature)
    matplotlib.pyplot .ylabel("count")
    matplotlib.pyplot .title('Categorical data')
    matplotlib.pyplot .figure(figsize=(3,3))
    matplotlib.pyplot .show
```

```python
#  5.	Study the occurrence of CVD across different ages.

pandas.crosstab(Cleaned_dataframe.age,Cleaned_dataframe.target).plot(kind="bar",figsize=(20,20))
matplotlib.pyplot.title('occurrence of CVD across different ages')
matplotlib.pyplot.xlabel('Age')
matplotlib.pyplot.ylabel('Range')
matplotlib.pyplot.legend(["Healthy", "Disease"])
```

```python
#  6.	Can we detect heart attack based on anomalies in resting blood pressure of the patient?

pandas.crosstab(Cleaned_dataframe.trestbps,Cleaned_dataframe.target).plot(kind="bar",figsize=(20,20))
matplotlib.pyplot.title(' heart attack based on anomalies in resting blood pressure of the patient')
matplotlib.pyplot.xlabel(' resting blood pressure')
matplotlib.pyplot.ylabel('Range')
matplotlib.pyplot.legend(["Healthy", "Disease"])
```

#### Blood pressure is not an accurate predictor of a heart attack. Sometimes a heart attack can cause an increase or decrease in blood pressure, but having a change in blood pressure reading doesn't always mean it's heart-related. Instead, a better strategy for gauging a heart attack is to look at your overall symptoms.

```python
# 12.	Perform logistic regression, predict the outcome for test data, and validate the results by using the confusion matrix.

x=Cleaned_dataframe.iloc[:,0:14]
y=Cleaned_dataframe.iloc[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
x=std.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30)
```

```python
from sklearn.linear_model import LogisticRegression
LGR=LogisticRegression()
LGR.fit(x_train,y_train)
Array=LGR.predict(x_test)
```

```python
Array
```

```python
test=y_test.tolist()
```

```python
final=pandas.DataFrame(list(zip(test,Array)),columns=['Given_Target','Prediction_Target'])
final
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,Array))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Array,y_test)
```
